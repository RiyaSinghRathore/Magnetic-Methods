#### <span style="font-family: Times New Roman;">1. Develop a function to determine, whether a given number is even or odd.</span>

```python
def function1(n):
    if n%2==0: return "EVEN" 
    else: return "ODD"
```

```python
# Example
function1(235)
```

#### <span style="font-family: Times New Roman;">2. Develop a function to search the Maximum and Minimum from an Array of Numbers.</span>

```python
def function2(arr):
    max = min = arr[0] 
    for n in arr:
        if n > max : max = n  
        elif n < min: min = n     
    return max, min
```

```python
# Example
Array = [5, 3, 8, 1, 9, 2, 7]
max, min = function2(Array)
print("Maximum :", max)
print("Minimum :", min)
```


#### <span style="font-family: Times New Roman;">3. Plot the given topography data with proper labeling.</span>

```python
# Importing neccesary libraries

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation

import pandas as pd
```

```python
data = pd.read_table("Data/Topography_practical_1.txt", sep='\\s+')
print(data)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(data['X'], data['Y'], data['Z'], c=data['Z'], cmap='terrain')
plt.colorbar(scatter, label='Z')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Scatter Contour Plot of Typography Data')
plt.grid()
```

```python
# Create and save the GIF
def update(angle): ax.view_init(elev=10, azim=angle)
ani = FuncAnimation(fig, update, frames=range(0, 360, 5))
ani.save('Plots/P1_3DContour.gif', writer='imagemagick', fps=30)
```
